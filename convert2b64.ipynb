```python
from PIL import Image
from io import BytesIO
import base64
from os.path import exists
from tqdm.notebook import tqdm, trange
tqdm.pandas()
```

```python
data_root = '/raid/AISSEL/Hamed/datasets/caption_data'
```

```python
image_root = '/raid/AISSEL/Hamed/datasets/wit/images'
```

```python
def b64(uniq_id):
    fn = image_root + '/' + uniq_id + '.jpg'
    base64_str = None
    if exists(fn):
        try:
            img = Image.open(fn)
            img_buffer = BytesIO()
            img.save(img_buffer, format=img.format)
            byte_data = img_buffer.getvalue()
            base64_str = str(base64.b64encode(byte_data))[2:-1]
        except:
            base64_str = None
            pass
    return base64_str
```

```python
ofa_all = f'/home/barzamini/text_processing/ofa.tsv'

df_wit = pd.read_csv(ofa_all, sep='\t')
df_wit['image_id'] = df_wit["uniq_id"].values.tolist()
df_wit['predicted object labels'] = ""
```

```python
df_wit['image'] = df_wit["uniq_id"].progress_apply(b64).values.tolist()
```
