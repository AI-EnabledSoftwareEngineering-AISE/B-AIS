```python
image_root = '/raid/AISSEL/Hamed/datasets/wit/images'
```

```python
def b64(uniq_id):
    fn = image_root + '/' + uniq_id + '.jpg'
    base64_str = None
    if exists(fn):
        try:
            img = Image.open(fn)
            img_buffer = BytesIO()
            img.save(img_buffer, format=img.format)
            byte_data = img_buffer.getvalue()
            base64_str = base64.b64encode(byte_data)
        except:
            base64_str = None
            pass
    return base64_str
```

```python
ofa_all = f'/home/barzamini/text_processing/ofa.tsv'

df_wit = pd.read_csv(ofa_all, sep='\t')
df_wit['image_id'] = df_wit["uniq_id"].values.tolist()
df_wit['predicted object labels'] = ""
```
